```python
#28/08/2023 - scanned pdf reading try
import fitz  # PyMuPDF
import pytesseract
from PIL import Image
from docx import Document
from io import BytesIO

def extract_images_from_pdf(pdf_path):
    images = []
    doc = fitz.open(pdf_path)
    for page_num in range(doc.page_count):
        page = doc.load_page(page_num)
        images += page.get_pixmap()
    doc.close()
    return images

def perform_ocr(image):
    img = Image.frombytes("RGB", [image.width, image.height], image.samples)
    text = pytesseract.image_to_string(img)
    return text

def create_word_document(text, output_path):
    doc = Document()
    doc.add_paragraph(text)
    doc.save(output_path)

def convert_scanned_pdf_to_word(pdf_path, output_path):
    scanned_images = extract_images_from_pdf(pdf_path)
    extracted_text = []
    for image in scanned_images:
        text = perform_ocr(image)
        extracted_text.append(text)
    
    combined_text = "\n".join(extracted_text)
    create_word_document(combined_text, output_path)

pdf_path = "E:\\Project Code\\10_Process_Complete_pdf\\95.pdf"
output_path = "output.docx"
convert_scanned_pdf_to_word(pdf_path, output_path)



```
